### Introduction

* Open source (i.e., Original code is avaible for free and it can be shared and modified), Object oriented (classes, inheritance, objects, and overloading), High level programming language(Closer to Human language).
* By a Dutch programmer `Guido Van Rossum` in late 1980s.
* Interpreted language(i.e., Checks line by line). Therefore prototyping is done quickly
* Portable (Works on different platforms i.e., Windows, Mac, Linux, Raspberry Pi, etc.)
* Extendable (Can add a built-in functions,modules, packages written in C/C++/Java)
* Numerous Libraries & Great support.
* Uses: Web development, GUIs (Graphical User Interfaces), Games, Electronic design automation, Database access etc.
* Softwares: IDLE, VS Code, Anaconda, PyCharm etc.
* CPython - Python implementation on standard C language.
* Jython - Python implementation with Java virtual machine to blend with Java.
* Pypy - Python implemented in Python and its Just-in time compiler making it fastest.
* Iron Python - for windows, which implements common runtime libraries to interface with .NET.

###  Variables

* Memory location to store a value (Value can be string, numeric, etc)
* Variable is assigned to value before referencing and it can accessed, updated later.
* Rules:
    - Should begin with a letter or underscore.
    - Should never begin with a number.
    - Should never be a reserved word.
    - Case sensitive.
    - Tip: Short and cute names will be better
* Data type of a variable can be known with the `type()` function.
* Casting can be done to specify the data type of a variable.

````python
x=2
y="yes"
w = 99.9
print(int(w))

````

````python
A=99.9
print(int(A))
````
